```python
import json
import time
from tqdm import tqdm

purchases = {}

with open('data\\purchase_log.txt', 'r', encoding='utf-8') as f:
    for line in tqdm(f):

        dict_ = json.loads(line)
        purchases.update({dict_['user_id']: dict_['category']})

with open('data\\funnel.csv', 'w', encoding='utf-8') as newf:

    with open('data\\visit_log.csv', 'r', encoding='utf-8') as f:
        
        for line in tqdm(f):
            user_id = line.strip().split(',')[0]
            source = line.strip().split(',')[1]
            if user_id in purchases:
                newf.write(f'{user_id},{source},{purchases[user_id]}\n')
            else:
                newf.write(f'{user_id},{source}\n')

```
